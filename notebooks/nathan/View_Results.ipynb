```python
nc_res_df = pd.read_csv('./results/node_class.csv')
ep_res_df = pd.read_csv('./results/edge_pred.csv')
```

This notebook is used to view the results tables generated by Generate_Results.ipynb

# Node Classification Results

```python
nc_res_df = nc_res_df[nc_res_df['Embed'].isin(['LM','TAPE','default'])]
nc_res_df.groupby(by=['Dataset','Agg','Embed'])['Acc'].max().reset_index()\
    .sort_values(by=['Dataset','Agg','Acc'],ascending=False)
```

```python
view = {
    'dataset':'Cora',
    'aggregator':'pool',
    'pca':True,
}
nc_res_view = nc_res_df[(nc_res_df['Dataset']==view['dataset'])\
                        &(nc_res_df['Agg']==view['aggregator'])\
                        &(nc_res_df['PCA']==view['pca'])]
nc_res_view.groupby(by=['Dataset','Agg','Embed'])['Acc'].max().reset_index()\
    .sort_values(by=['Dataset','Agg','Acc'],ascending=False)
```

# Edge Prediction Results

```python
ep_res_df = ep_res_df[ep_res_df['Embed'].isin(['LM','TAPE','default'])]
ep_res_df.groupby(by=['Dataset','Agg','Embed'])['AUC'].max().reset_index()\
    .sort_values(by=['Dataset','Agg','AUC'],ascending=False)
```

```python
view = {
    'dataset':'Cora',
    'aggregator':'pool',
    'pca':True,
}
ep_res_view = ep_res_df[(ep_res_df['Dataset']==view['dataset'])\
                        &(ep_res_df['Agg']==view['aggregator'])\
                        &(ep_res_df['PCA']==view['pca'])]
ep_res_view.groupby(by=['Dataset','Agg','Embed'])['Max'].max().reset_index()\
    .sort_values(by=['Dataset','Agg','Max'],ascending=False)
```
